#Introducción

Mucho se ha hablado de la ley de atracción ente fuerzas: cargas opuestas se atraen y cargas del mismo signo se repelen. Este principio es el que se emplea para el funcionamiento del Motor Electrostático que consiste en el movimiento giratorio de un objeto sobre su propio eje impulsado por chispas eléctricas que generan dicho movimiento. La interacción entre las cargas se llama electrostática y es posible explicarla por medio de la Ley de Coulomb.

#Marco teórico

Los motores electrostáticos están basados en las fuerzas de atracción entre cargas eléctricas de distinto signo y las de repulsión entre cargas del mismo signo. El motor electrostático de Franklin consiste esencialmente en una rueda sin llanta que gira en el plano horizontal sobre cojinetes con un bajo coeficiente de rozamiento.
Motor electrostático o motor del condensador es un tipo de motor eléctrico de acuerdo con la atracción y la repulsión de carga eléctrica. Requiere típicamente una fuente de tensión, aunque los motores muy pequeños emplean tensiones más bajas. Los motores eléctricos convencionales emplean la atracción y la repulsión magnéticas, y requieren de gran intensidad en las bajas tensiones. 
En 1750, los primeros motores electrostáticos fueron desarrollados cerca Benjamín Franklin y Andrew Gordon. El motor electrostático encuentra hoy uso frecuente en micro-mecánico (MEMS) los sistemas donde están sus voltajes de la impulsión debajo de 100 voltios, y donde moviendo las placas cargadas son lejos más fáciles de fabricar que bobinas y corazones de hierro. También, la maquinaria molecular que funciona las células vivas se basa a menudo en los motores electrostáticos lineares y rotatorios.


#Desarrollo del experimento y resultados: 

Queremos generar el movimiento de una esfera haciendo uso de dos cargas, $Q_{1}$ que es positiva y $Q_{2}$ que es negativa. 

Iremos desarrollando paso por paso la manera en como podriamos obtener este fenómeno físico y que es lo que necesitamos para ello.

Primero definamos las variables de nuestro experimento:

Para facilitar los calculos tendremos una esfera de radio $R = 1m$ y masa $m = 0.5kg$















```
#Figura 1
Image ('melectrostatico.PNG',width= 750,height=400)
```

Con estas variables y conociendo que es una esfera hueca podemos conocer el momento de inercia: 

$$I = \frac{2}{3}mR^2$$

Reemplazando nuestra ecuación obtenemos el momento de inercia 

$$I = \frac{2}{3}(0.5)(1)$$

$$I = \frac{1}{3} kg * m^2$$

Esta esfera posee una carga en su superficie una carga $Q_{3} = 0.3$C la cual es inducida por otras cargas.

Lo siguiente es tomar dos cargas, una positiva y otra negativa que serán las responsables de darle movimiento a nuestra esfera. 

Definimos las cargas: 

$$Q_{1} = 1 C$$
$$Q_{2} = -1 C$$

La distancia entre la carga $Q_{1}$ y $Q_{3}$ es igual a la distancia entre $Q_{2}$ y $Q_{3}$ por lo que lo tomaremos como una misma constante, esta equivale a 

$$d = 1.3m$$

Pasemos ahora a calcular las fuerzas entre estas cargas, utilizando la Ley de Coulomb.


$$F = \frac{1}{4\pi\epsilon_{0}}\frac{Q*q}{d^2}$$




Reemplazando nuestras variables podemos calcular las fuerzas presentes: 

$$F_{1} = \frac{1}{4\pi\epsilon_{0}}\frac{Q_{1} Q_{3}}{d^2}$$

$$F_{2} = \frac{1}{4\pi\epsilon_{0}}\frac{Q_{2} Q_{3}}{d^2}$$

Ahora necesitamos hallar una relación entre estas fuerzas ejercidas y el movimiento que tendría la esfera, para ello haremos uso  de la 2da ley de Newton
$F = ma$ lo que en movimiento rotacional podemos expresar como $I\alpha = \tau$

Hallemos entonces los torques generados por estas fuerzas y sus distancia al centro de la esfera

$$\tau_{1} = F_{1} * R$$

Es importante analizar las direciones hacia las cuales actua nuestra fuerza par determinar los torques, en este caso nuestro $\tau_{2}$ tendria un valor negativo
$$\tau_{2} = - (F_{2} * R)$$


Despejando de la ecuación de la Ley de Newton $\alpha$ tenemos que 

$$\alpha = \frac{\tau_{1}+\tau_{2}}{I}$$

Ahora definimos nuestra velocidad inicial, como la esfera parte del reposo podemos tomar $\omega_{0} = 0$

Teniendo estos valores podemos saber la posicion de un punto en el extremo de nuestra esfera en un tiempo $t$

Como nuestra esfera posee una aceleración constante podemos definir el movimiento de la esfera como un movimiento circular uniformemente acelerado y ya que $\omega_{0} = 0$ nuestra ecuación quedaria de la siguiente manera 

$$\theta = \frac{1}{2}\alpha (t-t_{0})^2$$


Tomamos ahora un punto en todo que sea tangente a la esfera, que en el caso de nuestro experimento sería $Q_{3}$

La posición de $Q_{3}$ en $x$ esta dada por 

$$Pos_{x} = R*cos{\theta}$$

Y en $y$ su posición es

$$Pos_{y} = R * \sin\theta$$


Por ultimo despejamos $\theta$ de nuestra ecuación para obtener la posición de $Q_{3}$ en un tiempo $t$

$$Pos_{x} = R* \cos(\frac{1}{2}\alpha (t-t_{0})^2)$$

$$Pos_{y} = R* \sin(\frac{1}{2}\alpha (t-t_{0})^2)$$

#Conclusiones
- Los campos eléctricos tienen su origen en diferencias de voltaje: entre más elevado sea el voltaje, más fuerte será el campo que resulta. Campos magnéticos tienen su origen en los corrientes eléctricos: un corriente más fuerte resulta en un campo más fuerte.
-Un campo eléctrico existe, aunque no haya corriente. Cuando hay corriente, la magnitud del campo magnético cambiará con el consumo de poder, pero la fuerza del campo eléctrico quedará igual.
-Desde el punto de vista de la mecánica, pudimos revisar el concepto de momento de una fuerza respecto de un punto, como producto de dos magnitudes: el módulo de la fuerza y el brazo o distancia de la dirección de la fuerza a dicho punto.
