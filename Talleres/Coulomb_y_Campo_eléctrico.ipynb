En cada esquina de un cubo cuyo lado es l existe una carga puntual

```
#Figura 1
Image ('cube.PNG',width= 200,height=200)
```

##Solución: 

Lo primero que debemos hacer es enumerar estas cargas de 1 a 8 de la siguiente manera.


```
#Figura 2 
Image('cube1 (1).png', width=200, height=200)
```

Luego procedemos a definir la fuerzas que actúan sobre la carga situada en el origen $Q_{1}$.

Primero tomamos las cargas que se encuentran a una distancia $l$ de la carga $Q_{1}$

$(1)$
$$\vec{F_{12}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}\hat{i}$$ 

$(2)$
$$\vec{F_{13}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}\hat{j}$$ 

$(3)$
$$\vec{F_{14}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}\hat{k}$$ 

Ahora calculamos las cargas que se encuentran diagonal a la carga $Q_{1}$ de la siguiente manera.

Tenemos dos lados con longitud $l$, por el Teorema de Pitágoras podemos definir que la diagonal entre los puntos máximos de estos lados es igual a $D =\sqrt{l^2 + l^2}$ lo que es igual a $l \sqrt{2}$.

Tomamos las cargas que se encuentran a una distancia de $l \sqrt{2}$ de la carga $Q_{1}$.  

$(4)$
$$\vec{F_{15}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{2l^2}(\frac{1}{\sqrt{2}}\hat{i}+ \frac{1}{\sqrt{2}}\hat{j} )$$ 


$(5)$
$$\vec{F_{16}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{2l^2}(\frac{1}{\sqrt{2}}\hat{i}+ \frac{1}{\sqrt{2}}\hat{k} )$$ 


$(6)$
$$\vec{F_{17}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{2l^2}(\frac{1}{\sqrt{2}}\hat{j}+ \frac{1}{\sqrt{2}}\hat{k} )$$ 

Por último tomamos la carga $Q_{8}$ que se encuenta en la esquina opuesta superior del cubo, para calcular la distancia hacemos lo siguiente: 

Formamos un triángulo rectángulo que va de la carga $Q_{1}$ hasta carga $Q_{8}$, dando como resultado un triángulo de base $l \sqrt{2}$ , altura $l$ y la hipotenusa es igual a la distancia entre las cargas, para hallar esta distancia recurrimos al Teorema de Pitágoras $\sqrt{(l\sqrt{2})^2 + l^2}$ dando como resultado: $l\sqrt{3}$

Por lo que el vector de fuerza sería: 

$(7)$
$$\vec{F_{18}}= \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{3l^2}(\frac{1}{\sqrt{3}}\hat{i}+ \frac{1}{\sqrt{3}}\hat{j}+ \frac{1}{\sqrt{3}}\hat{k})$$ 


Lo siguiente es sumar los componentes de los vectores: 

Suma de componentes en $\hat{i}$

$\vec{F_{12}} + \vec{F_{15}} + \vec{F_{16}} + \vec{F_{18}} = \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}(1 + \frac{2}{2 \sqrt{2}}+ \frac{1}{3\sqrt{3}}) $

Resultado: 

$\vec{F_{12}} + \vec{F_{15}} + \vec{F_{16}} + \vec{F_{18}} = \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}(1.90) $

Suma de componentes en $\hat{j}$

$\vec{F_{13}} + \vec{F_{15}} + \vec{F_{17}} + \vec{F_{18}} = \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}(1 + \frac{2}{2 \sqrt{2}}+ \frac{1}{3\sqrt{3}}) $

Resultado: 

$\vec{F_{13}} + \vec{F_{15}} + \vec{F_{17}} + \vec{F_{18}} = \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}(1.90)$

Suma de componentes en $\hat{k}$

$\vec{F_{14}} + \vec{F_{16}} + \vec{F_{17}} + \vec{F_{18}} = \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}(1 + \frac{2}{2 \sqrt{2}}+ \frac{1}{3\sqrt{3}}) $

Resultado: 

$\vec{F_{14}} + \vec{F_{16}} + \vec{F_{17}} + \vec{F_{18}} = \frac{1}{4\pi \epsilon_{0}}\frac{Q^2}{l^2}(1.90)$


Por ultimo calculamos la magnitud del vector resultante: 

$|F| = \frac{Q^2}{\epsilon_{0} l^2} \sqrt{(\frac{1.90}{4\pi})^2+(\frac{1.90}{4\pi})^2+(\frac{1.90}{4\pi})^2} $

Como resultado obtenemos que la fuerza que actúa sobre la carga en el origen 0 debido a las demás cargas es: 

R// $$F = \frac{0.265Q^2}{\epsilon_{0}l^2}$$

##Problema 2: 

Se dobla una varilla que tiene poco espesor para formar el arco de un
círculo cuyo radio es R, la varilla transporta una carga uniforme por unidad
de longitud l. El arco contiene un ángulo total de 2θ , que es simétrico al
eje x, como se muestra en la figura. Determine el campo eléctrico E en el
origen 0.


```
#Figura 3 
Image('problem2.png', width=200, height=200)
```

#Solución: 

Primero debemos dividir esta distribución de carga uniforme en partes discretas de carga $Q$,
Para ello tomamos pequeños incrementos del ángulo $dθ$ de la siguiente manera.


```
#Figura 4
Image('problem2(1).PNG', width=200, height=200)
```

Tomamos este $dQ$ como una densidad de carga lineal.

Despejando tendremos entonces que la carga en ese pequeño diferencial es
 (1)$$dQ = lRd\theta$$ 

Ahora debemos aplicar la Ley de Coulomb para obtener el campo eléctrico producido por este $dQ$ de la siguiente manera: 

$$k = \frac{1}{4\pi \epsilon_{0}}$$
(2)
$$dE = k\frac{dQ}{R^2}(-\cos \theta \hat{i} - \sin \theta \hat{j})$$

El siguiente paso es sumar todos estos pequeños campos eléctricos formados para obtener el campo eléctrico total

$$E = \int dE$$
Reemplazando $dE$ (Ecuación (2)) y sacando las constantes de la integral tenemos que

 $$E = \frac{k}{R^2} \int dQ(-\cos \theta \hat{i} -sin \theta \hat{j})$$ 

Lo siguiente es reemplazar este $dQ$ (Ecuación (1)) y establecer nuestros límites de integración que van desde $+\theta_{0}$ hasta $-\theta_{0}$ según nuestra gráfica.

$$E = \frac{k}{R^2} \int_{-\theta_{0}}^{+\theta_{0}} lR (-\cos \theta \hat{i}- \sin \theta)d\theta$$

Sacamos el signo - como factor común y junto cn las constantes lo situamos fuera de la integral, dando como resultado lo siguiente: 

$$E = - \frac{kl}{R}[\hat{i}\int_{-\theta_{0}}^{+\theta_{0}} \cos{\theta} d\theta + \hat{j}\int_{-\theta_{0}}^{+\theta_{0}} \sin{\theta} d\theta]$$

Luego de realizar las integrales obtenemos que la componente en $\hat{j}$ se cancela, entonces nos queda lo siguiente: 

$$E = -\frac{kl}{R}\hat{i}2\sin{\theta_{0}}$$

Ordenando la ecuación tenemos como resultado: 

R// $$E = - \frac{2kl\sin{\theta_{0}}}{R} \hat{i}$$

##Problema 3 

Una carga positiva q se coloca en el centro de un anillo circular de radio R. El
anillo transporta una carga negativa que está distribuida en forma uniforme cuya
magnitud total es –Q. (a) Si la carga q se desplaza desde el centro a una distancia
pequeña x como se muestra en la figura, demuestre que experimentará un
movimiento armónico simple una vez que se haya liberado. (b) Si su masa es m,
¿cuál será su periodo?



```
#Figura 5
Image('problem3.PNG', width=200, height=200)
```

##Solución:

Primero partimos de la ecuación de intensidad de campo eléctrico

$$E = \frac{F}{q}$$

Despejando la fuerza obtenemos 

$$ F = qE$$

Lo siguiente es remplazar el campo eléctrico de este anillo con carga distribuida uniformemente el cual está dado por: 

$$E =  \frac{1}{4\pi \epsilon_{0}} \frac{Qx}{(x^2 + R^2)^{3/2}}$$

Reemplazando y sabiendo que nuestro anillo esta cargado por $-Q$ obtenemos lo siguiente

$$ F = \frac{q}{4\pi \epsilon_{0}}\frac{-Qx}{(x^2 + R^2)^{3/2}}$$

Podemos ver que $x$ es mucho menor que el radio del anillo $R$ por lo que podemos dejar simplemente $(R^2)^{3/2}$ en el denominador de nuestra fracción, dando como resultado  

$$F = -\frac{qQ}{4\pi \epsilon_{0}R^3}x$$

Ahora, esta ecuación luce bastante similar a la Ley de Hooke, así que podemos tomar todo lo que multiplica a la $x$ como la constante de elasticidad $k$, de esta manera:

a)
$$k = \frac{qQ}{4\pi\epsilon_{0} R^3}$$

De la anterior ecuación podemos deducir que tenemos una fuerza restauradora que es proporcional a su desplazamiento, por lo que podemos deducir que se presenta un movimiento armónico simple.

Para dar respuesta a la pregunta b) establecemos primero la fórmula para el período $T$ sabiendo que tenemos una masa $m$

$$T = 2\pi\sqrt{\frac{m}{k}}$$

Conocemos el valor de $k$ por lo que la reemplazamos en la ecuación anterior de la siguiente manera

$$T =2\pi \sqrt{\frac{m4\pi\epsilon_{0}R^3}{qQ}}$$

De la raíz podemos sacar un 2 al operar el 4 y una $R$ al sacar raíz cuadrada a $R^3$ por lo que ordenando la ecuación quedaría así

b)
$$T = 4\pi R\sqrt{\frac{m\pi\epsilon_{0}R}{qQ}}$$

Ordenamos los valores que necesitamos para resolver el problema: 

Constante elástica de un resorte ideal
$$ k = 126 N/m$$

Masa de la esfera
$$ m = 0.800kg$$

Al halar la bala hacia abajo le damos una amplitud
$$A = 0.05m$$

Carga eléctrica
$$Q = -3x10^{-6}$$

Nuestra esfera oscilará sinusoidalmente alrededor del punto de equilibrio, con amplitud $A$.

La frecuencia angular de la esfera esta dada por la siguiente fórmula

$$\omega = \sqrt{\frac{k}{m}}$$

Reemplazando obtenemos que

$$\omega = \sqrt{\frac{126}{0.8}}$$

Operando la ecuación

$$\omega = 12.5$$ 

La distancia entre la esfera y la tabla estará dada por la distancia inicial entre ambas menos el cambio en la posición de la esfera debido a su movimiento armónico simple

$$ r = 0.150 - 0.05cos{12.5t}$$

$$E = \frac{1}{4\pi\epsilon_{0}} \frac{|Q|}{r^2}$$

Reemplazamos los valores y tomando $\frac{1}{4\pi\epsilon_{0}} =8.988 * 10^9$ obtenemos lo siguiente

$$E = \frac{(8.988 * 10^9)(3 * 10 ^{-6})}{[0.150 - 0.050\cos(12.5t)]^2}$$

Realizamos la multiplicación del numerador

$$E = \frac{2.70 * 10^4}{[0.150 - 0.050\cos(12.5t)]^2}$$

Podemos sacar del denominador como factor común el 0.05,

$$E = \frac{2.70 * 10^4}{[(0.05)(0.3 - \cos(12.5t))]^2}$$

Por último simplificamos la ecuación y obtenemos la intensidad del campo eléctrico en función de $t$

##Problema 6 

Calcule el campo eléctrico en el origen para la distribución lineal
de carga que se muestra en la figura. La parte izquierda (rojo) tiene
una carga +Q y la parte derecha (azul ) una carga -Q.





```
#Figura 8
Image('problema6.PNG',width=200, height=200)
```

Lo primero que debemos hacer es determinar la expresión del diferencial de la magnitud del campo eléctrico de la siguiente manera

$$k = \frac{1}{4\pi\epsilon_{0}}$$

$$dE_{p} = \frac{kdq}{a^2}$$

(En la gráfica el radio $R$ esta expresado como $a$)


Expresemoslo vectorialmente

$$d\vec{E_{p}} = d{E_{x}}\hat{i} + dE_{y}\hat{j}$$

Pasemos a determinar las componentes en x y en y de este vector, para eso debemos considerar un diferencial de carga de la varilla y este generará un campo eléctrico en el punto p, con la carga positiva el campo será radial y hacia afuera de la carga, formando un ángulo $\theta$ sobre la horizontal, en la carga negativa  se genera tambien un campo eléctrico en el punto p, pero con dirección hacia la carga, quedando de la siguiente manera


```
#Figura 9
Image('problem6(2).png', width=250, height=200)
```

Ambos diferenciales de carga $dq$ son iguales en magnitud y por tanto el diferencial de intensidad del campo eléctrico $dE$ también es igual, si analizamos las componentes en 'y' de la magnitud de la intensidad del diferencial del campo eléctrico estas se cancelan, ya que tienen la misma magnitud pero una dirección contraria. 

$$d\vec{E_{p}} = dE_{x}\hat{i} + 0\hat{j}$$

En x los componentes van en la misma dirección, por lo que tendremos dos veces este componente.

$$d\vec{E_{p}} = 2dE_{x}\hat{i}$$

Ahora pasemos a definir $dE_{x}$

$$dE_{x} = dE\cos\theta$$

Como ya conocemos $dE$ reemplazamos en la expresión, dando como resultado

$$d\vec{Ep} = \frac{2k\cos\theta dq}{a^2}\hat{i}$$

Como nuestra carga esta distribuida uniformemente tenemos que nuestra densidad lineal de carga es constante

$$\lambda = constante$$

Adaptando la fórmula de la densidad de carga lineal a nuestro problema tenemos que
$$\lambda = \frac{dq}{dl}$$

Despejando el diferencial de carga

$$dq = \lambda dl$$

$dl$ es la longitud de un diferencial de arco, por lo que podemos definirla así

$$dl = ad\theta$$

Reemplazando obtenemos que

$$dq = \lambda ad\theta$$

Lo siguiente es reemplazar este diferencial de carga en $d\vec{Ep}$

$$d\vec{Ep} = \frac{2k\cos\theta \lambda ad\theta}{a^2}\hat{i} $$

Pasemos ahora a realizar la integral respectiva

$$\vec{E_{p}} = \int_{\pi/2}^{\pi} \frac{2k\cos\theta \lambda ad\theta}{a^2}\hat{i}$$

Simplificando y sacando las constantes de la integral tenemos que 

$$\vec{E_{p}} = \frac{2k\lambda}{a}\int_{\pi/2}^{\pi} {\cos\theta d\theta}\hat{i}$$

Realizamos la integral y evaluación en los límites correspondientes,$\sin({\pi}) = 0$ por lo que nos queda lo siguiente 

$$\vec{E_{p}} = - \frac{2k\lambda}{a}\hat{i}$$

Ahora debemos escribir $\lambda$ en terminos de la carga de la figura.

$$\lambda = \frac{q}{l}$$

Reemplazamos la carga 

$$ \lambda = \frac{Q}{l}$$

Pero la longitud de arco $l$ es igual a 

$$l = a\theta$$

Reemplazando 

$$\lambda = \frac{Q}{a\theta}$$

Nuestra carga positiva (la carga negativa ya fue considerada en el procedimiento) va de $\pi /2$ a $\pi$ , ya que esta carga solo se encuentra en el segundo cuadrante tendremos que  $\theta = \pi /2 $

Nuestra expresión ordenada queda de la siguiente forma 

$$ \lambda = \frac{2Q}{\pi a}$$

Por último debemos sustituir $\lambda$ en la ecuación del vector del campo eléctrico 

$$\vec{E_{p}} = \frac{-2k}{a} \frac{2Q}{\pi a} \hat{i}$$

Simplificamos y obtenemos el campo eléctrico en el punto $p$ generado por la figura descrita en nuestro problema.

R//
$$ \vec{E_{p}} = \frac{-4kQ}{\pi a^2} \hat{i}$$
